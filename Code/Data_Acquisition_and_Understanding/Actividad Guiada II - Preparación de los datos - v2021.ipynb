### Construcción de atributos

**Atributo 1 a generar:** edad de los alumnos

```python
# Construcción de atributos - edad del alumno

def calcularEdad(row):
    anio_nac = int(row['fecha_nac_alumno'])
    return 2021 - anio_nac


edad = personas.apply(lambda row: calcularEdad(row), axis=1).rename('edad')
edad
```

```python
# Se agrega el atributo al dataset de datos_persona

personas = pd.concat([personas, edad], axis=1)
personas.head(1)
```

```python
# Se comienza con la unificación de los tres datasets

temp = pd.merge(personas, academicos, on='nro_inscripcion', how='inner')
temp.head()
```

**Atributo 2 a generar:** años de inactividad entre egreso del nivel medio e ingreso a nivel superior de los alumnos

```python
# Antes de comenzar se debería revisar la cantidad de nulos
nulos_ingreso = temp.fecha_ingreso_alumno.isna().sum()
nulos_egreso = temp.fecha_egr_sec.isna().sum()

print(f"Cantidad de filas con valores nulos en atributo fecha_ingreso_alumno: {nulos_ingreso}")
print(f"Cantidad de filas con valores nulos en atributo fecha_egr_sec: {nulos_egreso}")
```

```python
def calcularInactividad(row):
    egreso = int(row['fecha_egr_sec'])
    ingreso = int(row['fecha_ingreso_alumno'][-4:])
    return ingreso - egreso

inactividad = temp.apply(lambda row: calcularInactividad(row), axis=1).rename('inactividad')
inactividad
```

```python
# Se agrega el atributo al dataset temporal

temp = pd.concat([temp, inactividad], axis=1)
temp.head()
```

### Integración de los datos

```python
datos_completos = pd.merge(temp, censales, on='nro_inscripcion', how='inner')
datos_completos.head()
```

```python
# Se finaliza la eliminación de las columnas restantes

col_eliminar_completo = ['fecha_nac_alumno', 'fecha_egr_sec', 'fecha_ingreso_alumno', 'nro_inscripcion']

datos_completos.drop(col_eliminar_completo, inplace=True, axis=1)
datos_completos.head(1)
```

```python
# Se escribe a disco el dataset resultante

datos_completos.to_csv(r'../datasets/datos_completos.csv')
print(f"Cantidad de filas: {datos_completos.shape[0]}")
print(f"Cantidad de columnas: {datos_completos.shape[1]}")
```
